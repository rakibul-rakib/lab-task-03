```
 courses = []
choice = 'random'

def line():
    print('--------------------------------')

def show_menu():
    line()
    print('MENU')
    line()
    print('1. Add my semester Course')
    print('2. Unenroll any course')
    print('3. View all courses')
    print('4. Exit programme')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        course = input('Add any course:')
        courses.append(course)
        print('Course', course,'Course Added successfully!')
    elif choice == '2':
        if courses != []:
          print('Courses List:')
          for course in courses:
              print('●', course)
          course = input('Choice which course you want to unenroll: ')
          if course in courses:
              courses.remove(course)
              print('Successfully Unenroll course', course)
          else:
              print('Not found course', course)
        else:
              print('No course yet!')
    elif choice == '3':
        if courses != []:
          print('List of courses:')
          for course in courses:
              print('●', courses)
        else:
              print('No course yet')
    elif choice == '4':
        print('Good Bye!')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
